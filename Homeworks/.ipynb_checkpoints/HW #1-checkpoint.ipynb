```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import numpy as np
import scipy.stats as stats
import seaborn as sb

spamdata = pd.read_csv("spambase.csv") 
```

```python
plt.figure(figsize=(12,6))
plt.plot(spamdata['capitalLong'])
```

```python
freq = spamdata["type"].value_counts().sort_index()
freq.plot.bar(title="Frequency of Spam and Nonspam")
```

```python
spamdata.plot.scatter(x='num3d',y='num000')
```

```python
spamdata.hist(column='money',bins=40,)
```

```python
freq.plot.pie(autopct='%1.1f%%')
```

```python
summary = spamdata['capitalAve'].describe()
df = pd.DataFrame(summary)

styler = df.style
styler
```

```python
mean = statistics.mean(spamdata['our'])
median = statistics.median(spamdata['our'])
std = statistics.stdev(spamdata['our'])

print("mean: ")
print(mean)
print()
print("median:")
print(median)
print()
print("std:")
print(std)
```

```python
print("min:")
print(spamdata['hp'].min())
print()
print("max:")
print(spamdata['hp'].max())
print()
```

```python
percentile_25 = np.percentile(spamdata['telnet'], 25)
percentile_50 = np.percentile(spamdata['telnet'], 50)
percentile_75 = np.percentile(spamdata['telnet'], 75)


print("25th percentile:")
print(percentile_25)
print()
print("50th percentile:")
print(percentile_50)
print()
print("75th percentile:")
print(percentile_75)
print()
```

```python
mode = statistics.mode(spamdata['you'])

print("mode:")
print(mode)
```

```python
q3, q1 = np.percentile(spamdata['business'], [75 ,25])
iqr = q3 - q1

print("IQR:")
print(iqr)
```

```python
skewness = stats.skew(spamdata['credit'])
kurtosis = stats.kurtosis(spamdata['credit'], fisher=False)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
r = np.corrcoef(spamdata['capitalAve'], spamdata['capitalTotal'])

print("Correlation Coefficient: ", r[0,1])
```

```python
range = spamdata['num857'].max() - spamdata['num857'].min()

print("range: ", range)
```

```python
corr_matrix = spamdata.corr(numeric_only=True)

print(corr_matrix)
```

```python
dataplot = sb.heatmap(spamdata.corr(numeric_only=True), cmap="YlGnBu",)
```

```python
print(spamdata['remove'].corr(spamdata['internet']))
```

```python
print(spamdata['num3d'].corr(spamdata['num000']))
```

```python
spamdata.plot.scatter(x='font',y='money')
```

```python
print(spamdata['table'].corr(spamdata['technology']))
```

```python

#24 - Scatter plot is used for exploring relationships between two continuous variables. This is useful for spotting patterns, trends, clusters or correlations in the data—especially when each point is a unique observation.
```

```python

#25 - Pros: Works well to show proportions or percentages of a whole, particularly when you have a small number of categories and wish to emphasize their relative magnitude. Weakness: Hard to read with excessive classes or minor variation between the values; not useful for direct comparisons.

```

```python

#26 - This means that a histogram is used if you want to show a distribution, and thus bar chart is used if you want to compare things; however, line chart, rather like a histogram, is not a tool designed to compare things. Q.4 When you are plotting a time series, is it better to use a histogram version or the line version? How to choose between a histogram and a bar chart? (A) An answer is that when you are visualizing a time series (or other data with an associated order) and want to highlight trends, it is better suited to choose a line chart than a histogram. A histogram is used to describe the frequency distribution of a single continuous variable.
```

```python

#27 - You should use a table when you need to be more specific or if someone is looking to compare specific aspects of multiple dimensions. Best for numerical data where you need actual numbers to be treated better than graphical form.
```

```python
#28 - Histogram perfectly understands the distribution of continuous variable. It summarizes the number of data points that fall into a range of values (similar to bin), and is effective for visualizing the shape of the distribution of numerical data, to identify variances such as skewness, modality (unimodal, bimodal), spread, and outliers.

```
